```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


input_file = "output_time_flights.csv"
output_plot = "output_time_flights.svg"

column_names = ['algo','k','time']
df_all = pd.read_csv(input_file, header=None, names = column_names)

df_ = df_all.copy()
df_ = df_.pivot(index='k', columns='algo')
df_ = df_['time']


markers = ['^','o','*','s','x']
linestyles = [':','-.','--','-.','-']



#my_color = ['maroon', 'purple', 'red', 'green', 'sandybrown', 'dodgerblue']
my_color = ['black', 'black', 'black', 'black', 'black', 'black']

#df_ = df_[['SBID-SwapD','SBID-SwapI','SBID-Greedy','SBD','SBI']]
#df_ = df_.rename(columns={'iDiVE-Greedy':'i-DiVE-Greedy','iDiVE-SwapD':'i-DiVE-SwapD','iDiVE-SwapI':'i-DiVE-SwapI', 'Only Interestingness':'Only Importance'})
df_ = df_[['i-DiVE-SwapD','i-DiVE-SwapI','i-DiVE-Greedy','Only Diversity','Only Importance']]


markers = ['^','o','*','s','x']
linestyles = [':','-.','--','-.','-']


ax = df_.plot(kind='line', markersize=7, linewidth=1, color=my_color)
for i, line in enumerate(ax.get_lines()):
    line.set_marker(markers[i])

for i, line in enumerate(ax.get_lines()):
    line.set_linestyle(linestyles[i])

# for adding legend
ax.legend(ax.get_lines(), df_.columns, loc='best')
plt.xlabel(r'$\mathcal{k}$', fontsize=10)
plt.ylabel(r'$\mathcal{Time(s)}$', fontsize=10)
plt.xticks([5,15,25,35])
plt.savefig(output_plot, format="svg", dpi = 1000)
#ax.set_ylim(ymin=0)
plt.show()
```
