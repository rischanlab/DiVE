```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def to_percent(y, position):
    # Ignore the passed in position. This has the effect of scaling the default
    # tick locations.
    s = str(1.0 * y)

    # The percent symbol needs escaping in latex
    if matplotlib.rcParams['text.usetex'] is True:
        return s + r'$\%$'
    else:
        return s + '%'

input_file = "low_vs_high_query_load.csv"
output_plot = "low_vs_high_query_load.svg"

column_names = ['tradeoff','i-DiVE-Greedy-Adaptive-Pruning','i-DiVE-SwapD-Adaptive-Pruning']
df = pd.read_csv(input_file, header=None, names = column_names)

df = df[['tradeoff','i-DiVE-SwapD-Adaptive-Pruning','i-DiVE-Greedy-Adaptive-Pruning']]

df = df.set_index(['tradeoff'])

markers = ['o','+','^','*']
linestyles = ['-.','-',':','--']


#my_color = ['orange', 'magenta', 'maroon', 'red']

my_color = ['black', 'black', 'black', 'black', 'black', 'black']


ax = df.plot(kind='line', markersize=7, linewidth=1,color=my_color)
for i, line in enumerate(ax.get_lines()):
    line.set_marker(markers[i])

for i, line in enumerate(ax.get_lines()):
    line.set_linestyle(linestyles[i])

# for adding legend
ax.legend(ax.get_lines(), df.columns, loc='best')
formatter = FuncFormatter(to_percent)

# Set the formatter
plt.gca().yaxis.set_major_formatter(formatter)
plt.xlabel(r'$\lambda$', fontsize=10)
plt.ylabel('pruned queries gap', fontsize=10)
plt.autolayout : True
ax.set_ylim(ymax=50)
ax.set_ylim(ymin=0)
plt.savefig(output_plot, format="svg", dpi = 1000, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def to_percent(y, position):
    # Ignore the passed in position. This has the effect of scaling the default
    # tick locations.
    s = str(1.0 * y)

    # The percent symbol needs escaping in latex
    if matplotlib.rcParams['text.usetex'] is True:
        return s + r'$\%$'
    else:
        return s + '%'

input_file = "low_vs_high_new.csv"
output_plot = "low_vs_high_new.svg"

#column_names = ['tradeoff','i-DiVE-Greedy-Adaptive-Pruning','i-DiVE-SwapD-Adaptive-Pruning']
df = pd.read_csv(input_file, header=0)

#df = df[['tradeoff','i-DiVE-SwapD-Adaptive-Pruning','i-DiVE-Greedy-Adaptive-Pruning']]


df = df.set_index(['tradeoff'])
df = df.applymap(percentage_disease)

markers = ['x','*','>','o']
linestyles = [':','-','-.','--']


#my_color = ['orange', 'magenta', 'maroon', 'red']

my_color = ['black', 'black', 'black', 'black', 'black', 'black']


ax = df.plot(kind='line', markersize=7, linewidth=1,color=my_color)
for i, line in enumerate(ax.get_lines()):
    line.set_marker(markers[i])

for i, line in enumerate(ax.get_lines()):
    line.set_linestyle(linestyles[i])

# for adding legend
ax.legend(ax.get_lines(), df.columns, loc='best')
formatter = FuncFormatter(to_percent)

# Set the formatter
plt.gca().yaxis.set_major_formatter(formatter)
plt.xlabel(r'$\lambda$', fontsize=10)
plt.ylabel('number of pruned queries %', fontsize=10)
plt.autolayout : True
plt.legend(frameon=False)

ax.set_ylim(ymax=100)
ax.set_ylim(ymin=0)
plt.savefig(output_plot, format="svg", dpi = 1000, bbox_inches='tight')
plt.show()
```
